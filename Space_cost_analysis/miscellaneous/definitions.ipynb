Definitions by File

satellite_programs.csv — one row per program, the master reference table

-program_name — unique identifier for the spacecraft program, used to join all four files

-mission_type — the primary purpose of the satellite (Communications, Earth Observation, Navigation, Meteorology, Science, Military Surveillance, Technology Demo)

-orbit_class — the orbital regime (GEO, LEO, MEO, HEO, Interplanetary), which drives ops cost, maintenance intensity, and disposal approach

-dry_mass_kg — total spacecraft mass excluding propellant, in kilograms; the single strongest cost driver in most CER models

-bol_power_w — beginning-of-life power in watts; how much electrical power the solar arrays generate when newly launched

-design_life_yrs — how many years the spacecraft is designed to operate; drives reliability requirements and therefore cost

-num_instruments — number of payload instruments; drives payload complexity and software size

-software_sloc_k — software size in thousands of source lines of code; drives software development and maintenance cost

-contractor_heritage — how much prior experience the contractor has with similar systems (High/Medium/Low); high heritage reduces cost through reuse and learning

-technology_readiness_level — NASA/DoD TRL scale from 1–9; lower TRL means less mature technology and higher development risk and cost

-year_development_start — fiscal year when development officially began

-launch_year — fiscal year when the spacecraft launched and became operational

-end_of_life_year — fiscal year when the design life expires

-program_phase — current status as of 2025: Development, Operations, or Retired

-development_cost_fy2020_musd — total cost to design and develop the spacecraft, in millions of FY2020 dollars

-production_cost_fy2020_musd — total cost to manufacture and build the spacecraft hardware

-annual_ops_cost_fy2020_musd — average cost per year to operate the spacecraft (ground station staffing, tracking, mission operations)

-total_ops_cost_fy2020_musd — annual ops cost × design life; total operations spending over the full mission

-total_maintenance_cost_fy2020_musd — cumulative maintenance cost over the full mission life; increases year over year due to aging

-disposal_cost_fy2020_musd — cost to deorbit or move the spacecraft to a graveyard orbit at end of life

-total_lcc_fy2020_musd — total life cycle cost: development + production + total ops + total maintenance + disposal; the sum of everything
